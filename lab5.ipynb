```python
import numpy as np
import random
import time
from functools import wraps

def time_it(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        elapsed = end - start
        return result, elapsed
    return wrapper

def print_matrix(data):
    if isinstance(data, np.ndarray):
        data = data.tolist()
    if isinstance(data, list) and not isinstance(data[0], list):
        print(data)
        print()
        return

    for row in data:
        for elem in row:
            print(f"{elem:3}", end=" ")
        print()
    print()
```

## Завдання 1. Виконання вправ (Варіант 9)

**Вправа 1.** Вводиться число $n$. Вивести масив розміру $n \times n$, в якому по діагоналі йдуть числа від 1 до $n$, а інші числа рівні 0.

```python
@time_it
def task_1_np(n):
    return np.diag(np.arange(1, n + 1))

@time_it
def task_1_iter(n):
    arr = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == j:
                row.append(i + 1)
            else:
                row.append(0)
        arr.append(row)
    return arr

n = 5
res_np, time_np = task_1_np(n)
res_iter, time_iter = task_1_iter(n)

print(f"Numpy (час - {time_np:.6f})")
print_matrix(res_np)

print(f"Iter (час - {time_iter:.6f})")
print_matrix(res_iter)

res_np_l, time_np_l = task_1_np(1000)
res_iter_l, time_iter_l = task_1_iter(1000)

print(f"На великих значеннях:\nNumpy (час - {time_np_l:.6f})\nIter (час - {time_iter_l:.6f})")
print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

**Вправа 2.** Вводиться число $n$. Розставити 1 та 0 у шаховому порядку, починаючи з 1 в матриці розміром $n \times n$.

```python
@time_it
def task_2_np(n):
    arr = np.zeros((n, n), dtype=int)
    arr[::2, ::2] = 1
    arr[1::2, 1::2] = 1
    return arr

@time_it
def task_2_iter(n):
    arr = []
    for i in range(n):
        row = []
        for j in range(n):
            if (i + j) % 2 == 0:
                row.append(1)
            else:
                row.append(0)
        arr.append(row)
    return arr

n = 5
res_np, time_np = task_2_np(n)
res_iter, time_iter = task_2_iter(n)

print(f"Numpy (час - {time_np:.6f})")
print_matrix(res_np)

print(f"Iter (час - {time_iter:.6f})")
print_matrix(res_iter)

res_np_l, time_np_l = task_2_np(1000)
res_iter_l, time_iter_l = task_2_iter(1000)

print(f"На великих значеннях:\nNumpy (час - {time_np_l:.6f})\nIter (час - {time_iter_l:.6f})")
print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

**Вправа 4.** Вводяться числа $n$ і $m$. Вивести масив розміру $n \times m$, в якому у першому рядку йдуть числа від 0 до $m-1$, а всі інші елементи дорівнюють 0.

```python
@time_it
def task_4_np(n, m):
    arr = np.zeros((n, m), dtype=int)
    arr[0, :] = np.arange(m)
    return arr

@time_it
def task_4_iter(n, m):
    arr = []
    for i in range(n):
        row = []
        for j in range(m):
            if i == 0:
                row.append(j)
            else:
                row.append(0)
        arr.append(row)
    return arr

n, m = 4, 5
res_np, time_np = task_4_np(n, m)
res_iter, time_iter = task_4_iter(n, m)

print(f"Numpy (час - {time_np:.6f})")
print_matrix(res_np)

print(f"Iter (час - {time_iter:.6f})")
print_matrix(res_iter)

res_np_l, time_np_l = task_4_np(1000, 1000)
res_iter_l, time_iter_l = task_4_iter(1000, 1000)

print(f"На великих значеннях:\nNumpy (час - {time_np_l:.6f})\nIter (час - {time_iter_l:.6f})")
print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

**Вправа 7.** З клавіатури вводиться масив. Замінити всі нульові елементи на -1.

```python
@time_it
def task_7_np(data):
    arr = np.array(data)
    arr[arr == 0] = -1
    return arr

@time_it
def task_7_iter(data):
    res = []
    for x in data:
        if x == 0:
            res.append(-1)
        else:
            res.append(x)
    return res

data = [1, 0, 5, 0, 3, 0, 4]
res_np, time_np = task_7_np(data)
res_iter, time_iter = task_7_iter(data)

print(f"Numpy (час - {time_np:.6f})")
print(res_np)

print(f"Iter (час - {time_iter:.6f})")
print(res_iter)

data_l = data * 100000
res_np_l, time_np_l = task_7_np(data_l)
res_iter_l, time_iter_l = task_7_iter(data_l)

print(f"\nНа великих значеннях:\nNumpy (час - {time_np_l:.6f})\nIter (час - {time_iter_l:.6f})")
print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

**Вправа 8.** З клавіатури вводиться масив. Підрахувати в ньому кількість нульових та ненульових елементів.

```python
@time_it
def task_8_np(data):
    arr = np.array(data)
    zeros = np.count_nonzero(arr == 0)
    non_zeros = np.count_nonzero(arr != 0)
    return zeros, non_zeros

@time_it
def task_8_iter(data):
    z = 0
    nz = 0
    for x in data:
        if x == 0:
            z += 1
        else:
            nz += 1
    return z, nz

data = [1, 0, 5, 0, 3, 0, 4, 8, 9, 0]
res_np, time_np = task_8_np(data)
res_iter, time_iter = task_8_iter(data)

print(f"Numpy (час - {time_np:.6f})")
print(f"Нулів: {res_np[0]}, Не нулів: {res_np[1]}")

print(f"Iter (час - {time_iter:.6f})")
print(f"Нулів: {res_iter[0]}, Не нулів: {res_iter[1]}")

print(f"Співпадіння результатів: {np.allclose(np.array(res_np), np.array(res_iter))}\n")
```

**Вправа 12.** Створити масив одиниць розміром $n \times n$ та створити в ньому «рамку», сформовану з 0.

```python
@time_it
def task_12_np(n):
    arr = np.ones((n, n), dtype=int)
    arr[0, :] = 0
    arr[-1, :] = 0
    arr[:, 0] = 0
    arr[:, -1] = 0
    return arr

@time_it
def task_12_iter(n):
    arr = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                row.append(0)
            else:
                row.append(1)
        arr.append(row)
    return arr

n = 5
res_np, time_np = task_12_np(n)
res_iter, time_iter = task_12_iter(n)

print(f"Numpy (час - {time_np:.6f})")
print_matrix(res_np)

print(f"Iter (час - {time_iter:.6f})")
print_matrix(res_iter)

res_np_l, time_np_l = task_12_np(1000)
res_iter_l, time_iter_l = task_12_iter(1000)

print(f"На великих значеннях:\nNumpy (час - {time_np_l:.6f})\nIter (час - {time_iter_l:.6f})")
print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

**Вправа 16.** Заповнити вектор значеннями від 0 до $n$.Всі значення, що більші за $n/4$ та менші за $3n/4$ замінити нулями.

```python
@time_it
def task_16_np(n):
    arr = np.arange(n + 1)
    mask = (arr > n/4) & (arr < 3*n/4)
    arr[mask] = 0
    return arr

@time_it
def task_16_iter(n):
    arr = []
    lower = n / 4
    upper = 3 * n / 4
    for i in range(n + 1):
        if lower < i < upper:
            arr.append(0)
        else:
            arr.append(i)
    return arr

n = 10
res_np, time_np = task_16_np(n)
res_iter, time_iter = task_16_iter(n)

print(f"Numpy (час - {time_np:.6f})")
print(res_np)

print(f"Iter (час - {time_iter:.6f})")
print(res_iter)

print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

**Вправа 20.** Згенерувати вектор із $n$ випадкових елементів, що лежать в інтервалі (0,1). Замінити максимальний елемент на 0.

```python
@time_it
def task_20_np(data):
    arr = data.copy()
    arr[np.argmax(arr)] = 0
    return arr

@time_it
def task_20_iter(data):
    arr = data.tolist()
    
    n = len(arr)
    max_val = -1.0
    max_idx = -1
    
    for i in range(n):
        if arr[i] > max_val:
            max_val = arr[i]
            max_idx = i
    
    if max_idx != -1:
        arr[max_idx] = 0
    return arr

n = 5
input_data = np.random.rand(n)

print(f"Вхідні дані: {input_data}\n")

res_np, time_np = task_20_np(input_data)
res_iter, time_iter = task_20_iter(input_data)

print(f"Numpy (час - {time_np:.6f})")
print(res_np)

print(f"Iter (час - {time_iter:.6f})")
print(res_iter)

# Порівнюємо
print(f"Співпадіння результатів: {np.allclose(res_np, res_iter)}\n")
```

## Завдання 2. Розв'язання СЛАР (Варіант 9)

Згідно з варіантом 9, система має вигляд:
$$
\begin{cases}
x_1 + 2x_2 + x_3 + x_4 = 8 \\
2x_1 + x_2 + x_3 + x_4 = 5 \\
x_1 - x_2 + 2x_3 + x_4 = -1 \\
x_1 + x_2 - x_3 + 3x_4 = 10
\end{cases}
$$

Матричний вигляд $Ax=b$:
$$
A = \begin{pmatrix}
1 & 2 & 1 & 1 \\
2 & 1 & 1 & 1 \\
1 & -1 & 2 & 1 \\
1 & 1 & -1 & 3
\end{pmatrix}, \quad
b = \begin{pmatrix} 8 \\ 5 \\ -1 \\ 10 \end{pmatrix}
$$

```python
A = np.array([
    [1, 2, 1, 1],
    [2, 1, 1, 1],
    [1, -1, 2, 1],
    [1, 1, -1, 3]
], dtype=float)

b = np.array([8, 5, -1, 10], dtype=float)

def cramers_rule(A_in, b_in):
    n = A_in.shape[0]
    detA = np.linalg.det(A_in)
    
    if np.isclose(detA, 0):
        return None, detA
    
    x = np.zeros(n)
    for j in range(n):
        Aj = A_in.copy()
        Aj[:, j] = b_in
        x[j] = np.linalg.det(Aj) / detA
    return x, detA

x_cramer, det_main = cramers_rule(A, b)

b_calc = A @ x_cramer


A_inv = np.linalg.inv(A)
x_inv = A_inv @ b


x_solve = np.linalg.solve(A, b)

print(f"Визначник матриці A: {det_main:.4f}\n")
print(f"Розв'язок (Крамер): {x_cramer}")
print(f"Розв'язок (Inv):    {x_inv}")
print(f"Розв'язок (Solve):  {x_solve}\n")

print(f"Співпадіння результатів (Cramer vs Solve): {np.allclose(x_cramer, x_solve)}")
print(f"Співпадіння результатів (Inv vs Solve):    {np.allclose(x_inv, x_solve)}")
print(f"Перевірка A*x = b: {np.allclose(b_calc, b)}")
```

## Завдання 3. Обчислення матричного виразу (Варіант 9)

Обчислити значення виразу:
$$2A - (A^2 + B)B$$

Де:
$$A=\begin{pmatrix}1&4&2\\ 2&1&-2\\ 0&1&-1\end{pmatrix}, \quad B=\begin{pmatrix}4&6&-2\\ 4&10&1\\ 2&4&-5\end{pmatrix}$$

```python
A = np.array([[1, 4, 2],
              [2, 1, -2],
              [0, 1, -1]])

B = np.array([[4, 6, -2],
              [4, 10, 1],
              [2, 4, -5]])

@time_it
def task_3_numpy(A, B):
    return 2*A - (A @ A + B) @ B

@time_it
def task_3_iter(A_list, B_list):
    n = 3

    def mat_mul(X, Y):
        Z = [[0]*3 for _ in range(3)]
        for i in range(3):
            for j in range(3):
                for k in range(3):
                    Z[i][j] += X[i][k] * Y[k][j]
        return Z

    def mat_add(X, Y):
        return [[X[i][j] + Y[i][j] for j in range(3)] for i in range(3)]

    def mat_sub(X, Y):
        return [[X[i][j] - Y[i][j] for j in range(3)] for i in range(3)]

    def mat_scale(k, X):
        return [[k * x for x in row] for row in X]

    A_sq = mat_mul(A_list, A_list)
    Sum = mat_add(A_sq, B_list)
    Prod = mat_mul(Sum, B_list)
    TwoA = mat_scale(2, A_list)
    return mat_sub(TwoA, Prod)

res_np, t_np = task_3_numpy(A, B)
res_it, t_it = task_3_iter(A.tolist(), B.tolist())

print(f"NumPy (час - {t_np:.6f})")
print_matrix(res_np)

print(f"Iter (час - {t_it:.6f})")
print_matrix(np.array(res_it))

print(f"Співпадіння результатів: {np.allclose(res_np, res_it)}\n")
```
