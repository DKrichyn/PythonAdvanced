### Хід виконання
Реалізуємо функцію, яка стане основою декоратора `cache_decorator`. В тілі декоратора створюємо словник `memory`, який буде використовуватись як сховище (кеш) для результатів обчислень.

В обгортці `wrapper`, яка приймає `*args` та `**kwargs`, формується унікальний ключ на основі переданих аргументів функції.
1. Якщо цей ключ вже існує в `memory`, функція не виконується повторно, а результат повертається миттєво з повідомленням про використання кешу.
2. Якщо ключа немає, функція виконується, її результат записується в `memory`, після чого повертається.

Для коректної роботи з метаданими функцій використовуємо `@wraps` з модуля `functools`.

```python
from functools import wraps

def cache_decorator(func):
    memory = {}
    
    @wraps(func)
    def wrapper(*args, **kwargs):
        key = (args, frozenset(kwargs.items()))
        
        if key in memory:
            print(f"[{func.__name__}] Результат взято з кешу для аргументів: {args} {kwargs}")
            return memory[key]
        
        res = func(*args, **kwargs)
        memory[key] = res
        print(f"[{func.__name__}] Функцію виконано та закешовано для: {args} {kwargs}")
        return res
        
    return wrapper
```

### Реалізація тестових функцій
Створимо 4 функції для перевірки роботи декоратора з різними типами даних:

1.  `complex_calculation` — приймає два числа, імітує складне обчислення (піднесення до степеня).
2.  `string_concatenation` — працює з рядками (стрічками).
3.  `sum_many` — приймає довільну кількість аргументів через `*args`.
4.  `greet_user` — використовує іменовані аргументи `**kwargs`.

```python
@cache_decorator
def complex_calculation(a, b):
    return a ** b

@cache_decorator
def string_concatenation(str1, str2):
    return f"{str1} -> {str2}"

@cache_decorator
def sum_many(*args):
    return sum(args)

@cache_decorator
def greet_user(name, greeting="Hello"):
    return f"{greeting}, {name}!"
```

```python
print("=== Тест 1: Математична функція ===")
print(f"Result: {complex_calculation(2, 10)}") 
print(f"Result: {complex_calculation(2, 10)}") 
print(f"Result: {complex_calculation(3, 3)}") 
print("\n=== Тест 2: Робота з рядками ===")
print(string_concatenation("Python", "Lab"))
print(string_concatenation("Python", "Lab"))

print("\n=== Тест 3: Довільна кількість аргументів (*args) ===")
print(sum_many(1, 2, 3, 4, 5))
print(sum_many(1, 2, 3, 4, 5))
print(sum_many(10, 20))

print("\n=== Тест 4: Іменовані аргументи (**kwargs) ===")
print(greet_user(name="Dmytro", greeting="Hi"))
print(greet_user(name="Dmytro", greeting="Hi"))
print(greet_user(greeting="Hi", name="Dmytro"))
```
