### Умова завдання:

У введеному списку залишити тільки непрості числа і
відсортувати їх за зменшенням

### Приклади виконання:

In [2]: task_9( )<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;3 12 15 7 11 35 18

<span style="color:red">Out [2]:</span> [35, 18, 15, 12]

<br>

In [3]: task_9( )<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;0 34 2 1 15 22 4 23 6

<span style="color:red">Out [3]:</span> [34, 22, 15, 6, 4]

```python
def non_primes_desc():
    try:
        line = input("Введіть цілі числа через пробіл: ").strip()
        if not line:
            print("Помилка. Порожній ввід")
            return
        
        parts = line.split()
        nums = []
        for p in parts:
            try:
                nums.append(int(p))
            except ValueError:
                print(f"Помилка. Значення '{p}' має бути числовим")
                return
        
        def is_prime(n: int) -> bool:
            if n < 2:
                return False
            if n % 2 == 0:
                return n == 2
            d = 3
            while d * d <= n:
                if n % d == 0:
                    return False
                d += 2
            return True
        
        composite = [x for x in nums if x > 1 and not is_prime(x)]
        
        composite.sort(reverse=True)
        
        print(composite)
    
    except Exception as e:
        print("Помилка під час обробки:", e)
        return
```

```python
print("Тест 1:")
non_primes_desc()
```
